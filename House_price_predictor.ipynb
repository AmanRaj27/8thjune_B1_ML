Feature selection

```python
df.columns
```

```python
# Feature matrix
x=df[['Avg. Area Income','Avg. Area House Age','Avg. Area Number of Rooms','Area Population']]
# Label matrix
y=df['Price']
```

Data splitting

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=101)
```

```python
xtrain.info()
```

```python
xtest.info()
```

```python
ytrain.shape
```

```python
ytest.shape
```

Linear Regression Modelling

```python
from sklearn.linear_model import LinearRegression
```

```python
# Model creation and training
model=LinearRegression().fit(xtrain,ytrain)
```

```python
model
```

```python
model.coef_
```

```python
coef=pd.DataFrame(model.coef_,columns=['coeff'],index=x.columns)
coef
```

Model Evaluation

```python
round(100*model.score(xtest,ytest),2)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
ypred=model.predict(xtest)
```

```python
xtest.head()
```

```python
ypred[:5]
```

```python
ytest[:5].values
```

```python
print(mean_absolute_error(ytest,ypred))
```

```python
print(mean_squared_error(ytest,ypred))
```

```python
plt.scatter(ytest,ypred)
```

```python
pd.to_pickle(model,'house_price_predictor.pkl')
```
